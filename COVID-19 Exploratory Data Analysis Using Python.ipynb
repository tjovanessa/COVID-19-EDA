### 4. Penyelarasan kurva pertumbuhan seluruh negara dengan jumlah kasus kematian COVID-19 â‰¥ 25 (Bobot: 25%)

**Untuk mendapatkan gambaran perkembangan kasus covid-19 di berbagai negara, diperlukan penyelarasan kurva pertumbuhan setiap negara. Kurva hanya menampilkan informasi yang dimulai dengan hari dimana data kasus kematian COVID-19 setidaknya 25 orang.**

```python
deaths_growth = deaths_country.where(deaths_country.values >= 25)
deaths_growth = deaths_growth.T
deaths_growth = deaths_growth.count()
deaths_growth
```

Dari hasil di atas dapat dilihat bahwa dari data 77 hari, pada awal pencatatan tidak ada negara dengan kasus kematian COVID-19 setidaknya 25 orang. Sedangkan pada 7 April atau di akhir pencatatan, terdapat 57 negara yang memenuhi kualifikasi lebih dari atau sama dengan 25 kasus. Dapat disimpulkan bahwa jumlah kematian COVID-19 secara konstan meningkat.

**Visualisasikan hasil diatas dan berikan/atur judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot/kurva yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
# Plot time series
growth_case = deaths_growth
growth_case.plot(kind='line', figsize=(10, 5), color='r')
plt.title('Deaths Case by Country/Region')
plt.xlabel('Date')
plt.ylabel('Country Quantity')
plt.show()
```

Dari visualisasi plot di atas, ditunjukkan bahwa kasus kematian COVID-19 di berbagai negara yang berjumlah setidaknya 25 terjadi paling banyak di bulan April. Selama bulan Februari, jumlah negara dengan kasus 25 ke atas statis atau berjumlah sama. Mulai bulan Maret dan seterusnya, plot naik dan jumlah negara tidak pernah berkurang.

### 5. Visualisasi jumlah pasien yang sembuh (recovered people) (Bobot: 25%)
    
**Kita perlu mendapatkan informasi mengenai total pasien COVID-19 yang dinyatakan sembuh seperti pada kasus terkonfirmasi dan kasus kematian tapi, kali ini lakukan hal tersebut dengan membuat `user-defined functions` yang diperlukan bukan mengetik ulang barisan code yang sama seperti di nomor sebelumnya.**

```python
def recovered_data(recovered_country):
    recovered_country = recovered_country.groupby('Country/Region').sum()
    recovered_country = recovered_country.T
    recovered_country = recovered_country.drop(['Lat', 'Long'])
    return recovered_country

def recovered_info(recovered_country):
    return recovered_country.info()
```

**Periksa fungsi-fungsi yang sudah dibuat**

```python
recovered_data(raw_data_recovered)
```

```python
recovered_info(raw_data_recovered)
```

**Selanjutnya,tampilkan data total pasien COVID-19 yang sembuh dalam bentuk grafik yang sesuai. Berikan judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
recovered_country = raw_data_recovered
recovered_date = recovered_country.drop(['Lat', 'Long', 'Province/State', 'Country/Region'], axis = 1)
recovered_date.sum().plot(kind='line', figsize=(10, 5), color='g')
plt.title('Total Recovered Patient from COVID-19')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.show()
```

Dari informasi data dan visualisasi di atas, dapat disimpulkan bahwa kasus pasien sembuh dari COVID-19 dari awal bulan Januari 2020 hingga pertengahan Agustus 2021 sempat menurut 1x pada awal Desember 2020 namun secara stabil meningkat. Setelah Agustus 2021 dan seterusnya, jumlah pasien sembuh langsung turun ke angka 0 hingga akhir data.

**Lakukan visualisasi yang sama pada beberapa negara berikut (France, Spain, China, US, Italy, and Australia). Berikan judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
recovered_country = recovered_data(raw_data_recovered)
recovered_case = recovered_country[['France', 'Spain', 'China', 'US', 'Italy', 'Australia']]
recovered_case.plot(kind='line', figsize=(10, 5), alpha=1)
plt.title('Recovered Case by Country/Region')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.show()
```

Dari hasil visualisasi di atas, ditunjukkan bahwa Italia dan Amerika Serikat memiliki frekuensi tingkat kesembuhan yang tinggi, sesuai dengan banyaknya jumlah kasus terkonfirmasi. Puncak kasus sembuh dipegang oleh Amerika Serikat pada Desember 2021.
